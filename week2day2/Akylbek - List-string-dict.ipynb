# 📋 Part 1 — Lists

### What is a list?

A **list** is an ordered collection of items stored in a single variable. Think of it like a **row of numbered lockers**: each locker holds one item, and each has a number (its **position**) so you can find it again.

Instead of this:

```python
score1 = 90
score2 = 75
score3 = 88
```

…you keep everything in one place:

```python
scores = [90, 75, 88]
```

A list can hold numbers, strings, or a mix — and you can change it after creating it.

````python
# Creating lists
scores = [90, 75, 88]
names = ["Aru", "Daniyar", "Aigerim"]
mixed = [1, "two", 3.0]
empty = []          # a list with nothing in it yet

print(scores)
print(len(scores))  # len() tells you HOW MANY items: 3
````

### Indexing — getting one item out

Each item has a **position number** called its **index**. ⚠️ **Indexing starts at 0, not 1.** This is the #1 list trap.

```
 list:    ["Aru", "Daniyar", "Aigerim"]
 index:      0         1          2
```

- `names[0]` → `"Aru"` (the **first** item)
- `names[2]` → `"Aigerim"` (the **third** item)
- `names[-1]` → `"Aigerim"` (negative counts from the **end**; `-1` is the last item)

Asking for an index that doesn't exist (like `names[5]` here) gives an **IndexError**.

````python
names = ["Aru", "Daniyar", "Aigerim"]

print(names[0])    # Aru
print(names[2])    # Aigerim
print(names[-1])   # Aigerim  (last item)

# Lists are CHANGEABLE (mutable) — you can replace an item:
names[1] = "Bekzat"
print(names)       # ['Aru', 'Bekzat', 'Aigerim']
````

### Adding and removing items

| Action | Method | Example |
|---|---|---|
| Add to the end | `.append(x)` | `scores.append(100)` |
| Insert at a position | `.insert(i, x)` | `scores.insert(0, 50)` |
| Remove by value | `.remove(x)` | `scores.remove(75)` |
| Remove by position (and get it back) | `.pop(i)` | `last = scores.pop()` |
| How many items | `len(list)` | `len(scores)` |
| Is an item present? | `in` | `90 in scores` → `True` |

````python
scores = [90, 75, 88]

scores.append(100)     # add to end -> [90, 75, 88, 100]
scores.insert(0, 50)   # insert at front -> [50, 90, 75, 88, 100]
scores.remove(75)      # remove the value 75 -> [50, 90, 88, 100]
last = scores.pop()    # remove & return the last item (100)

print(scores)          # [50, 90, 88]
print("removed:", last)
print(90 in scores)    # True
````

### Slicing and looping

**Slicing** grabs a *range* of items with `list[start:stop]` (the `stop` index is **not** included):

```python
nums = [10, 20, 30, 40, 50]
nums[1:3]   # [20, 30]   -> positions 1 and 2
nums[:2]    # [10, 20]   -> from the start
nums[2:]    # [30, 40, 50] -> to the end
```

**Looping** over a list uses the `for` loop you already know:

````python
nums = [10, 20, 30, 40, 50]

# Loop over the items directly
for n in nums:
    print(n)

# Combine with functions + a loop: sum every item
def total(items):
    s = 0
    for item in items:
        s = s + item
    return s

print("Total:", total(nums))   # 150
````

### ✅ List cheat sheet

# 🔤 Part 2 — Strings

### What is a string?

A **string** is text: a sequence of characters wrapped in quotes. The best way to picture it is as a **list of letters** — everything you just learned about indexing, slicing, `len()`, and looping works on strings too.

```python
word = "Python"
```

```
 chars:  P  y  t  h  o  n
 index:  0  1  2  3  4  5
```

You can use single `'...'` or double `"..."` quotes — just be consistent.

````python
word = "Python"

print(len(word))    # 6
print(word[0])      # P  (indexing starts at 0, just like lists)
print(word[-1])     # n  (last character)
print(word[0:3])    # Pyt  (slice: stop index not included)

# Loop over characters
for letter in word:
    print(letter)
````

### ⚠️ Strings are IMMUTABLE

Unlike lists, you **cannot change a character in place**:

```python
word = "Python"
word[0] = "J"     # ❌ ERROR! Strings can't be modified this way.
```

Instead, you build a **new** string. String methods never change the original — they **return a new string**:

```python
word = "Python"
new_word = "J" + word[1:]   # "Jython"
```

This is the opposite of lists, which *can* be changed in place. Keep that contrast in mind.

### Useful string methods

| Method | What it does | Example | Result |
|---|---|---|---|
| `.upper()` | all uppercase | `"hi".upper()` | `"HI"` |
| `.lower()` | all lowercase | `"HI".lower()` | `"hi"` |
| `.strip()` | remove surrounding spaces | `"  hi  ".strip()` | `"hi"` |
| `.replace(a, b)` | swap text | `"cat".replace("c", "h")` | `"hat"` |
| `.split(sep)` | break into a list | `"a,b,c".split(",")` | `["a","b","c"]` |
| `.count(x)` | count occurrences | `"banana".count("a")` | `3` |
| `in` | is a substring present? | `"an" in "banana"` | `True` |


````python
text = "  Hello, World  "

print(text.strip())            # "Hello, World"
print(text.strip().upper())    # "HELLO, WORLD"  (methods can be chained)
print("a,b,c".split(","))      # ['a', 'b', 'c']
print("banana".count("a"))     # 3
print("an" in "banana")        # True

# f-strings: the clean way to build text with variables
name = "Aru"
age = 21
print(f"{name} is {age} years old.")   # Aru is 21 years old.
````

# 🗂️ Part 3 — Dictionaries

### What is a dictionary?

A **dictionary** (`dict`) stores data as **key → value** pairs. Think of a real dictionary: you look up a **word** (the key) to get its **definition** (the value). Or your phone contacts: a **name** (key) gives you a **number** (value).

Lists find things by **position** (`scores[0]`). Dictionaries find things by **name** (`phone["Aru"]`). That's the key difference.

```python
phone = {"Aru": "777-111", "Daniyar": "777-222"}
```

- `"Aru"` and `"Daniyar"` are **keys**.
- `"777-111"` and `"777-222"` are **values**.

````python
phone = {"Aru": "777-111", "Daniyar": "777-222"}

print(phone["Aru"])     # 777-111   (look up by key)
print(len(phone))       # 2          (number of pairs)

# Add or update a pair (same syntax)
phone["Aigerim"] = "777-333"   # add new
phone["Aru"] = "777-999"       # update existing
print(phone)
````

### ⚠️ The KeyError trap — use `.get()`

Looking up a key that doesn't exist **crashes** with a `KeyError`:

```python
phone["Bekzat"]   # ❌ KeyError — there's no such key
```

Safer way: `.get()` returns `None` (or a default you choose) instead of crashing:

```python
phone.get("Bekzat")              # None  (no crash)
phone.get("Bekzat", "unknown")   # "unknown"
```

To check if a key exists before using it, use `in` (it checks **keys**):

```python
"Aru" in phone     # True
```



### Looping over a dictionary

| You want… | Use |
|---|---|
| the keys | `for k in d:` or `d.keys()` |
| the values | `for v in d.values():` |
| both at once | `for k, v in d.items():` |


````python
phone = {"Aru": "777-111", "Daniyar": "777-222"}

# Loop over keys and values together
for name, number in phone.items():
    print(f"{name} -> {number}")

# Removing a pair
removed = phone.pop("Aru")    # removes "Aru" and returns its value
print("removed:", removed)
print(phone)
````

### ✅ Dictionary cheat sheet

```python
d = {"a": 1, "b": 2}
d["a"]              # look up by key (crashes if missing)
d.get("z", 0)       # safe look up with a default
d["c"] = 3          # add or update
d.pop("a")          # remove & return a value
len(d)              # number of pairs
"a" in d            # is this a KEY? (True/False)
d.keys(), d.values(), d.items()   # for looping
```

> Lists are indexed by **position** (0, 1, 2…). Dictionaries are indexed by **key** (a name you choose).
