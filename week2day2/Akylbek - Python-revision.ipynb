# 🧩 Functions in Python

### What is a function?

A **function** is a reusable block of code that does one job. Think of it like a **kitchen blender**:

- You put something **in** (fruit) → this is the **input**.
- The blender **does work** (blends it) → this is the **body** of the function.
- You get something **out** (a smoothie) → this is the **output**.

You don't rebuild the blender every time you want a smoothie. You build it once, then press the button whenever you need it. Functions work the same way: **write the code once, use it as many times as you want.**

### Why do we use functions?

1. **Avoid repetition** — write logic once instead of copy-pasting it everywhere.
2. **Organize code** — break a big problem into small, named pieces.
3. **Easier to fix** — if something is wrong, you fix it in one place.
4. **Easier to read** — `calculate_total()` tells you what the code does at a glance.

```python
# The simplest possible function

def say_hello():
    print("Hello, intern!")

# Nothing happens yet — we only DEFINED the function.
# To actually run it, we CALL it by writing its name with ():
say_hello()
say_hello()   # we can call it as many times as we want
```


### Breaking down the syntax

```python
def say_hello():
    print("Hello, intern!")
```

| Part            | Meaning                                                        |
|-----------------|----------------------------------------------------------------|
| `def`           | keyword that means "I am **def**ining a function"              |
| `say_hello`     | the **name** of the function (you choose it)                  |
| `()`            | where inputs go (empty here — no inputs)                       |
| `:`             | starts the function body                                       |
| indented lines  | the **body** — the code that runs when you call the function   |

> ⚠️ **Two-step rule:** Defining a function (`def ...`) does **not** run it. You must **call** it afterwards with `say_hello()`. This is the #1 thing beginners forget.


```python
# A function with an INPUT (called a parameter)

def greet(name):
    print("Hello, " + name + "!")

greet("Aigerim")
greet("Daniyar")
```

### Parameters and arguments

- A **parameter** is the placeholder you write in the definition: `def greet(name):` → `name` is a parameter.
- An **argument** is the real value you pass in when calling: `greet("Aigerim")` → `"Aigerim"` is an argument.

Simple way to remember: **parameter = the empty box**, **argument = what you put in the box**.

A function can take more than one parameter, separated by commas:


```python
# Multiple parameters

def add(a, b):
    result = a + b
    print(result)

add(3, 5)      # prints 8
add(10, 20)    # prints 30
```

### `return` — getting a value back

So far our functions only **print**. But usually we want the function to **give us a value back** so we can use it later. That's what `return` does.

```python
def add(a, b):
    return a + b
```

`return` sends the result **out** of the function. We can then store it, do math with it, or pass it to another function.


```python
def add(a, b):
    return a + b

# The returned value can be stored in a variable
total = add(3, 5)
print(total)          # 8

# ...or used directly in more math
print(add(2, 2) * 10)  # 40

# ...or even passed into another function call
print(add(add(1, 1), add(2, 2)))  # add(2, 4) -> 6
```


### 🚨 The most important beginner trap: `print` vs `return`

These look similar but are **completely different**.

| | `print()` | `return` |
|---|---|---|
| What it does | shows text on the screen | sends a value back to your code |
| Can you reuse the value later? | ❌ No | ✅ Yes |
| Who is it for? | the **human** reading the output | the **rest of your program** |

```python
def add_print(a, b):
    print(a + b)      # only shows it

def add_return(a, b):
    return a + b      # gives it back

x = add_print(2, 3)   # screen shows 5, but x is None!
y = add_return(2, 3)  # screen shows nothing, but y is 5

print(x)   # None  <- nothing was returned
print(y)   # 5     <- the value came back to us
```

**Rule of thumb:** if you only want to *see* something → use `print`. If you want to *use* the result later → use `return`.


```python
# Default parameter values
# If the caller doesn't provide a value, the default is used.

def greet(name, greeting="Hello"):
    return greeting + ", " + name + "!"

print(greet("Aigerim"))                 # Hello, Aigerim!
print(greet("Daniyar", "Good morning")) # Good morning, Daniyar!
```



### Default parameters & a quick note on scope

- **Default parameter:** giving a parameter a value in the definition (`greeting="Hello"`) makes it optional. If the caller skips it, the default is used.
- **Scope (light version):** variables created **inside** a function live only inside that function. They disappear when the function finishes. So a variable named `result` inside `add()` is invisible to the outside world. This is good — it keeps functions self-contained and prevents accidental clashes.


```python
# Putting it ALL together with what you already know:
# numbers + if/else + loops, wrapped in a function with a return value.

def grade(score):
    if score >= 90:
        return "A"
    elif score >= 75:
        return "B"
    elif score >= 60:
        return "C"
    else:
        return "F"

# Use a loop to test the function on several scores
for s in [95, 80, 65, 40]:
    print(s, "->", grade(s))
```


### ✅ Cheat sheet

```python
def function_name(parameter1, parameter2="default"):
    # body: the work happens here
    result = parameter1 + parameter2
    return result          # send a value back (optional)

value = function_name(10, 5)   # call it, store the result
```

Remember:
- **`def`** to define, **`()`** to call.
- **Parameter** = box in the definition, **argument** = value you pass in.
- **`print`** shows to the human; **`return`** gives the value back to your code.
- Variables inside a function stay inside the function.
